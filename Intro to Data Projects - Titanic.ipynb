```python
import pandas as pd
from pandas import Series,DataFrame

# set up Titanic csv file as DataFrame
titanic_df = pd.read_csv('train.csv')

titanic_df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.factorplot("Sex",data=titanic_df,aspect=1, kind="count",color='b')
```

```python
sns.factorplot('Sex',data=titanic_df,aspect=1,kind='count',hue='Pclass')
```

```python
sns.factorplot('Pclass',data=titanic_df,aspect=1,kind='count',hue='Sex')
```

```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
    
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
titanic_df[0:10]
```

```python
sns.factorplot('Pclass',data=titanic_df,aspect=1,kind='count',hue='person')
```

```python
fig = sns.FacetGrid(titanic_df,aspect=4,hue='person',)

fig.map(sns.kdeplot,'Age',shade=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df,aspect=4,hue='Pclass')

fig.map(sns.kdeplot,'Age',shade=True)

oldest = titanic_df['Age'].max()

fig.set(xlim=(0,oldest))

fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
```

```python
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']

sns.factorplot('Cabin',data=cabin_df,aspect=1,kind='count',palette='winter_d')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != "T"]
sns.factorplot('Cabin',data=cabin_df,aspect=1,kind='count',palette='summer')
```

```python
# who was alone and who was with family
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
titanic_df['Alone']
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df.head()
```

```python
sns.factorplot('Alone',data=titanic_df,palette='Blues',aspect=1,kind='count')
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0:'No', 1:'Yes'})
sns.factorplot('Survivor',data=titanic_df,palette='Set1',aspect=1,kind='count')
```
